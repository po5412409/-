```python
import pandas as pd
from scipy import stats

df = pd.read_csv('../data/dataset/코스피_최종.csv')
df.head(2)
```

```python
df.describe()
```

```python
# 타겟 변수에 따라 두 그룹으로 나눔
group0 = df[df['분식기업'] == 0]
group1 = df[df['분식기업'] == 1]
```

```python
df.columns[6:]
```

## T-Test 진행

```python
# 결과 저장할 리스트
results = []

# 독립 변수들에 대해 t-test 진행
for col in df.columns[6:]:
    stat, p_value = stats.ttest_ind(group1[col], group0[col], nan_policy='omit')  # t-test 수행
    results.append({'Variable': col, 't-statistic': stat, 'p-value': p_value})

# 결과를 데이터프레임으로 변환
t_test_results = pd.DataFrame(results)
```

```python
# p-value가 유의미한 변수만 필터링 (예: 0.05 이하)
t_test_feature = t_test_results[t_test_results['p-value'] < 0.05]
t_test_feature
```

## 비모수 검정 (Mann-Whitney U test)

```python
from scipy.stats import mannwhitneyu

# 결과 저장할 리스트
results2 = []

# 각 컬럼에 대해 Mann-Whitney U 테스트 수행
for column in df.columns[6:]:
    stat, p_value = mannwhitneyu(group1[column], group0[column])
    results2.append({'Variable': column, 't-statistic': stat, 'p-value': p_value})

# 결과를 데이터프레임으로 변환
whitney_test_results = pd.DataFrame(results2)
```

```python
# p-value가 유의미한 변수만 필터링 (예: 0.05 이하)

whitney_test_feature = whitney_test_results[whitney_test_results['p-value'] < 0.05]
whitney_test_feature
```

```python
t_test_variable = t_test_feature[['Variable']]
whitney_test_variable = whitney_test_feature[['Variable']]

# 두 컬럼을 이어붙여 C라는 새로운 데이터프레임으로 생성
compare_df = pd.concat([t_test_variable, whitney_test_variable], axis=1)
compare_df
```
