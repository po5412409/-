## 필요한 컬럼 연결재무제표를 통해 생성

```python
df = pd.read_csv('../data/dataset/코스피_연결재무제표계정값.csv')
```

```python
df.columns
```

### 1. 안정성 중 유동비율, 당좌비율 컬럼 생성
- 유동비율 : 유동자산 / 유동부채
- 당좌비율 : 당좌자산 / 유동부채

```python
# 유동비율
kospi['유동비율(%)'] = round((df['유동자산(천원)']/df['유동부채(천원)'])*100, 2)

# 당좌비율
kospi['당좌비율(%)'] = round((df['당좌자산(천원)']/df['유동부채(천원)'])*100, 2)
```

### 2. 활동성 중 매출채권회전율, 재고자산회전율, 고정자산회전율 컬럼 생성

- 매출채권회전율 : 당기 매출액 / 평균매출채권
- 재고자산회전율 : 당기 매출액 / 평균재고자산
- 고정자산회전율 : 당기 매출액 / 평균고정자산

```python
# (당기+전기)평균매출채권
for company in df['Name'].unique():
    company_data = df[df['Name'] == company]
    company_data['평균매출채권'] = (company_data['매출채권(천원)'] + company_data['매출채권(천원)'].shift(1)) / 2 
    df.loc[company_data.index, '평균매출채권'] = company_data['평균매출채권']

# (당기+전기)평균재고자산
for company in df['Name'].unique():
    company_data = df[df['Name'] == company]
    company_data['평균재고자산'] = (company_data['재고자산(천원)'] + company_data['재고자산(천원)'].shift(1)) / 2 
    df.loc[company_data.index, '평균재고자산'] = company_data['평균재고자산']

# (당기+전기)평균비유동자산
for company in df['Name'].unique():
    company_data = df[df['Name'] == company]
    company_data['평균비유동자산'] = (company_data['비유동자산(천원)'] + company_data['비유동자산(천원)'].shift(1)) / 2 
    df.loc[company_data.index, '평균비유동자산'] = company_data['평균비유동자산']
```

```python
# 매출채권회전율 값 채우기
kospi['매출채권회전율'] = round(df['매출액(천원)'] / df['평균매출채권'], 2)

# 재고자산회전율 값 채우기
kospi['재고자산회전율'] = round(df['매출액(천원)'] / df['평균재고자산'], 2)

# 고정자산회전율 값 채우기
kospi['고정자산회전율'] = round(df['매출액(천원)'] / df['평균비유동자산'], 2)
```

### 3. 현금흐름 중 영업현금흐름/투자현금흐름 파생변수 생성

```python
df.info()
```

```python
# 영업현금흐름/투자현금흐름 파생변수 생성
kospi['영업현금흐름/투자현금흐름(%)'] = round(df['영업활동으로인한현금흐름(천원)'] / df['투자활동으로인한현금흐름(천원)'], 2)
```
