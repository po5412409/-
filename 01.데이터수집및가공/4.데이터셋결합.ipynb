코스피 재무비율 + Beneish + Benford

```python
import pandas as pd

kospi = pd.read_csv('../data/dataset/코스피_재무비율.csv')
beneish = pd.read_csv('../data/dataset/코스피_beneish.csv')
benford = pd.read_csv('../data/dataset/코스피_benford.csv')
label = pd.read_csv('../data/dataset/코스피_label.csv')
```

```python
df = pd.merge(kospi, beneish, on=['Symbol', 'Name', '회계년', '결산월', '주기'])
df = pd.merge(df, benford, on=['Symbol', '회계년', '결산월', '주기'])
```

```python
# df[df['Name_x'] != df['Name_y']]
df.rename(columns={'Name_x' : 'Name'}, inplace = True)
df.drop(columns = 'Name_y', inplace=True)
```

```python
df = pd.merge(df, label, on=['Name', '회계년'], how='left')
```
