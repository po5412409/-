```python
import pandas as pd

kospi_profitability = pd.read_csv('../data/raw_data/코스피/코스피_수익성.csv', thousands=',')
kospi_growth = pd.read_csv('../data/raw_data/코스피/코스피_성장성.csv', thousands=',')
kospi_stability = pd.read_csv('../data/raw_data/코스피/코스피_안정성.csv', thousands=',')
kospi_activity = pd.read_csv('../data/raw_data/코스피/코스피_활동성.csv', thousands=',')
kospi_cf = pd.read_csv('../data/raw_data/코스피/코스피_현금흐름.csv', thousands=',')
# kospi_beneish = pd.read_csv('./raw_data/코스피/코스피_beniesh_M-score.csv', thousands=',')
# kospi_benford = pd.read_csv('./raw_data/kospi_benford.csv')
# kospi_company = pd.read_csv('./감리지적기업연도/코스피_분식기업.csv')
```
