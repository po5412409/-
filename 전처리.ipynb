```python
# 쉼표를 제거하고 숫자로 변환
active['매출액(천원)'] = active['매출액(천원)'].str.replace(',', '').astype(float)
active['매출채권(천원)'] = active['매출채권(천원)'].str.replace(',', '').astype(float)
active['재고자산(천원)'] = active['재고자산(천원)'].str.replace(',', '').astype(float)
```

```python
# (당기+전기)평균매출채권
for company in active['Name'].unique():
    company_data = active[active['Name'] == company]
    company_data['평균매출채권'] = (company_data['매출채권(천원)'] + company_data['매출채권(천원)'].shift(1)) / 2 
    active.loc[company_data.index, '평균매출채권'] = company_data['평균매출채권']

# (당기+전기)평균재고자산
for company in active['Name'].unique():
    company_data = active[active['Name'] == company]
    company_data['평균재고자산'] = (company_data['재고자산(천원)'] + company_data['재고자산(천원)'].shift(1)) / 2 
    active.loc[company_data.index, '평균재고자산'] = company_data['평균재고자산']
```

```python
# 매출채권회전율 값 채우기
active['매출채권회전율'] = active['매출액(천원)'] / active['평균매출채권']

# 재고자산회전율 값 채우기
active['재고자산회전율'] = active['매출액(천원)'] / active['평균재고자산']
```

```python
# Extracting the list of financial company symbols from the kospi financial list
financial_symbols = kospi_financial_companies.iloc[:, 0].values

# Function to filter and save CSVs
def filter_and_save(df, financial_symbols, filename):
    filtered_df = df[~df['Symbol'].isin(financial_symbols)]
    filtered_df = df[~df['Name'].str.contains('은행|증권|투자|스팩|보험|화재|금융|카드|지주|신탁')]
    filtered_df.to_csv(f'코스피_{filename}', index=False, encoding = 'utf-8-sig')

# Applying filter to each dataset and saving new CSVs
filter_and_save(kospi_beniesh_m_score, financial_symbols, '1beniesh_M-score.csv')
filter_and_save(kospi_growth, financial_symbols, '1성장성.csv')
filter_and_save(kospi_profitability, financial_symbols, '1수익성.csv')
filter_and_save(kospi_stability, financial_symbols, '1안정성.csv')
filter_and_save(kospi_cash_flow, financial_symbols, '1현금흐름.csv')
filter_and_save(kospi_activity, financial_symbols, '1활동성.csv')

# Confirming files saved
saved_files = [
    '코스피_beniesh_M-score.csv',
    '코스피_성장성.csv',
    '코스피_수익성.csv',
    '코스피_안정성.csv',
    '코스피_현금흐름.csv',
    '코스피_활동성.csv'
]
saved_files
```
