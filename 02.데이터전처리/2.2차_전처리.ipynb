### 4. 연결재무제표 계정값으로 계산 이후 결측치 값들 개별제무제표 계정값으로 계산 후 대체

```python
df.isnull().sum()
```

```python
# 해당 컬럼 결측치 데이터
print(df[df['총자본증가율(전년동기)(%)'].isnull()][['Name' , '회계년']])
print(df[df['영업이익증가율(전년동기)(%)'].isnull()][['Name' , '회계년']])
print(df[df['당기순이익증가율(전년동기)(%)'].isnull()][['Name' , '회계년']])
print(df[df['DSRI'].isnull()][['Name' , '회계년']])
```

---
- 총자본증가율(전년동기)(%)
- 영업이익증가율(전년동기)(%)
- 당기순이익증가율(전년동기)(%)
- DSRI

```python
value = pd.read_csv('../data/dataset/코스피_개별재무제표계정값.csv')
value = pd.merge(df, value, on = ['Symbol', 'Name', '결산월', '회계년', '주기'], how = 'left')

value['영업이익(천원)_전기'] = value.groupby('Name')['영업이익(천원)'].shift(1)
value['당기순이익(천원)_전기'] = value.groupby('Name')['당기순이익(천원)'].shift(1)
value['자본총계_전기'] = value.groupby('Name')['총자본(천원)'].shift(1)
value['매출액(천원)_전기'] = value.groupby('Name')['매출액(천원)'].shift(1)
value['매출채권(천원)_전기'] = value.groupby('Name')['매출채권(천원)'].shift(1)
```

```python
# 총자본증가율: 자본총계[당기]/자본총계[전기]-1
df['총자본증가율(전년동기)(%)'] = df['총자본증가율(전년동기)(%)'].mask(df['총자본증가율(전년동기)(%)'].isna(),
                                                   round(((value['총자본(천원)']-value['자본총계_전기'])/value['자본총계_전기'])*100, 2))

# 영업이익증가율: (영업이익[당기]-영업이익[전기])/영업이익[전기]
df['영업이익증가율(전년동기)(%)'] = df['영업이익증가율(전년동기)(%)'].mask(df['영업이익증가율(전년동기)(%)'].isna(),
                                       round(((value['영업이익(천원)']-value['영업이익(천원)_전기'])/value['영업이익(천원)_전기'])*100, 2))

# 순이익증가율: (당기순이익[당기]-당기순이익[전기])/당기순이익[전기]
df['당기순이익증가율(전년동기)(%)'] = df['당기순이익증가율(전년동기)(%)'].mask(df['당기순이익증가율(전년동기)(%)'].isna(),
                                       round(((value['당기순이익(천원)']-value['당기순이익(천원)_전기'])/value['당기순이익(천원)_전기'])*100, 2))

# DSRI: (매출채권[당기] / 매출액[당기]) / (매출채권[전기] / 매출액[전기])
df['DSRI'] = df['DSRI'].mask(df['DSRI'].isna(),
                                       round(((value['매출채권(천원)']/value['매출액(천원)'])/(value['매출채권(천원)_전기']/value['매출액(천원)_전기'])), 2))
```

### 6. 무한대 값 처리

- 매출채권회전율
- 재고자산회전율
- DSRI


모든 결측치가 label = 0인 기업이기 때문에
- 무한대 값 (inf/-inf) -> 결측치 -> 제거

매출채권회전율, 재고자산회전율 먼저 개별재무제표 계정값으로 계산시도

```python
import numpy as np

df.isin([np.inf, -np.inf]).sum()
```

```python
# 매출채권회전율 inf값 총 9개
df[np.isinf(df['매출채권회전율'])]
```

```python
value = pd.read_csv('../data/dataset/코스피_개별재무제표계정값.csv')

# (당기+전기)평균매출채권
for company in value['Name'].unique():
    company_data = value[value['Name'] == company]
    company_data['평균매출채권'] = (company_data['매출채권(천원)'] + company_data['매출채권(천원)'].shift(1)) / 2 
    value.loc[company_data.index, '평균매출채권'] = company_data['평균매출채권']

# (당기+전기)평균재고자산
for company in value['Name'].unique():
    company_data = value[value['Name'] == company]
    company_data['평균재고자산'] = (company_data['재고자산(천원)'] + company_data['재고자산(천원)'].shift(1)) / 2 
    value.loc[company_data.index, '평균재고자산'] = company_data['평균재고자산']

value = pd.merge(df, value, on = ['Symbol', 'Name', '결산월', '회계년', '주기'], how = 'left')
```

```python
# 매출채권회전율: 매출액 / 평균매출채권
df['매출채권회전율'] = df['매출채권회전율'].mask(np.isinf(df['매출채권회전율']),
                                                   round((value['매출액(천원)'] / value['평균매출채권']), 2))

# 재고자산회전율 = 매출액 / 평균재고자산
df['재고자산회전율'] = df['재고자산회전율'].mask(np.isinf(df['재고자산회전율']),
                                                   round((value['매출액(천원)'] / value['평균재고자산']), 2))
```

---
- 매출채권회전율은 처리 완
- 재고자산회전율은 개별로 처리 불가
- 재고자산회전율 inf값 전부 0으로 대체

```python
# 재고자산회전율 컬럼에서 inf 값을 0으로 대체
df['재고자산회전율'] = df['재고자산회전율'].replace([np.inf, -np.inf], 0)
```
