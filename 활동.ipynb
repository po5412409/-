```python
import pandas as pd

kospi = pd.read_csv('../진짜자료/코스피_합체.csv')
df = pd.read_csv('../datasets/코스피코스닥/코스피_beniesh_M-score.csv')
```

```python
kospi
```

```python
df
```

```python
act =pd.DataFrame()
act['Symbol'] = df['Symbol']
act['Name'] = df['Name']
act['회계년'] = df['회계년']
act['매출원가'] = df['매출원가(천원)']
act['재고자산'] = df['재고자산(천원)']
act
```

```python
# 쉼표를 제거하고 숫자로 변환
act['매출원가'] = act['매출원가'].str.replace(',', '').astype(float)
act['재고자산'] = act['재고자산'].str.replace(',', '').astype(float)
```

```python
act
```

```python
act['재고자산'].isna().sum()
```

```python
for company in act['Name'].unique():
    company_data = act[act['Name'] == company]
    company_data['평균재고자산'] = (company_data['재고자산'] + company_data['재고자산'].shift(1)) / 2 
    act.loc[company_data.index, '평균재고자산'] = company_data['평균재고자산']
```

```python
act['평균재고자산'].isna().sum()
```

```python
act.head(20)
```

```python
for company in act['Name'].unique():
    company_data = act[act['Name'] == company]
    company_data['재고자산회전율'] = (company_data['매출원가'] + company_data['평균재고자산'])
    act.loc[company_data.index, '재고자산회전율'] = company_data['재고자산회전율']
act
```

```python
act['재고자산회전율'].isna().sum()
```

```python
act['재고자산회전율'] = act['재고자산회전율'].round(2)
act.head(20)
```

```python
kospi = pd.merge(kospi, act, on = ['Symbol', 'Name', '회계년'], how = 'inner')
```

```python
act.to_csv('코스피_재고자산회전율.csv', index = False, encoding='utf-8-sig')
```
