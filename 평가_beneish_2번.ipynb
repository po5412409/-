```python
import pandas as pd
import numpy as np

# 데이터 로드
data = pd.read_csv('datasets/코스피.csv')

# 타겟 변수 및 피처 선택
X = data[['당기순이익률(%)', '총자본증가율(전년동기)(%)', '무형자산구성비율(%)', '타인자본비율(%)', '당좌비율(%)', '매출채권회전율(회)', '영업현금흐름/투자현금흐름(%)', 'SGAI']].copy()
y = data['분식기업'].copy()

# M_score 계산
X['tl'] = (0.929 + 1.024 * X['당기순이익률(%)'] + 0.833 * X['총자본증가율(전년동기)(%)'] + 1.350 * X['무형자산구성비율(%)'] + 
                1.166 * X['타인자본비율(%)'] + 0.948 * X['당좌비율(%)'] + 0.771 * X['매출채권회전율(회)'] + 
                1.085 * X['영업현금흐름/투자현금흐름(%)'] + 0.996 * X['SGAI']).round(2)

# 'manipulate' 열을 NaN으로 초기화
X['manipulate'] = np.nan

# 조건에 따라 'manipulate' 열 채우기
# X.loc[X['tl'] < -1.39, 'manipulate'] = 0
# X.loc[X['tl'] >= -1.39, 'manipulate'] = 1

# 임곗값 0.5
X.loc[X['tl'] < 0.91632, 'manipulate'] = 0
X.loc[X['tl'] >= 0.91632, 'manipulate'] = 1

# 실제 분식 회계 기업과 예측 비교
X['real_mani'] = y
result = X[(X['manipulate'] == 1) & (X['real_mani'] == 1)]

# 결과 확인
print(result)
```
