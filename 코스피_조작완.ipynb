```python
import pandas as pd
import numpy as np

# CSV 파일 불러오기
df = pd.read_csv('./전처리/개별재무제표/코스닥_베네시.csv')

# NaN 값과 무한대 값 처리
#df.replace([np.inf, -np.inf], np.nan, inplace=True)  # 무한대 값을 NaN으로 변환
#df.fillna(0, inplace=True)  # 모든 NaN 값을 0으로 대체

# 쉼표 제거 및 실수형 변환
columns_to_convert = ['매출액(천원)', '매출채권(천원)', '매출원가(천원)', '유동자산(천원)', '재고자산(천원)',
                      '유형자산(천원)', '총자산(천원)', '판매비와관리비(천원)', '당기순이익(천원)', '유동부채(천원)',
                      '비유동부채(천원)', '영업활동으로인한현금흐름(천원)', '영업이익(천원)', '유무형자산상각비(천원)']
for column in columns_to_convert:
    df[column] = df[column].astype(str).str.replace(',', '').astype(float)

# 데이터 확인
print(df.head())
```

```python
# 회사명으로 데이터 그룹화하여 DSRI 계산
for company in df['Name'].unique():
    company_data = df[df['Name'] == company]
    company_data['DSRI'] = (company_data['매출채권(천원)'] / company_data['매출액(천원)']) / (company_data['매출채권(천원)'].shift(1) / company_data['매출액(천원)'].shift(1))
    df.loc[company_data.index, 'DSRI'] = company_data['DSRI']
```

```python
# t-1년도와 t년도의 값을 비교하여 새로운 열을 만들거나 해당 열을 변경
# # {(전기 매출액 - 전기 매출원가) / 전기 매출액} / {(당기 매출액 - 당기 매출원가) / 당기 매출액}
# GMI
for company in df['Name'].unique():
    company_data = df[df['Name'] == company]
    company_data['GMI'] = ((company_data['매출액(천원)'].shift(1) - company_data['매출원가(천원)'].shift(1)) / company_data['매출액(천원)'].shift(1)) / ((company_data['매출액(천원)'] - company_data['매출원가(천원)']) / company_data['매출액(천원)'])
    df.loc[company_data.index, 'GMI'] = company_data['GMI']
```

```python
# t년도와 t-1년도의 값을 비교하여 새로운 열을 만들거나 해당 열을 변경
# ((1 - 당기유동자산 + 당기유형자산) / 당기총자산) / ((1 - 전기유동자산 + 전기 유형자산) / 전기총자산)
# AQI
for company in df['Name'].unique():
    company_data = df[df['Name'] == company]
    company_data['AQI'] = (1 - (company_data['유동자산(천원)'] + company_data['유형자산(천원)']) / company_data['총자산(천원)'])  / (1 - (company_data['유동자산(천원)'].shift(1) + company_data['유형자산(천원)'].shift(1)) / company_data['총자산(천원)'].shift(1))
    df.loc[company_data.index, 'AQI'] = company_data['AQI']
```

```python
# t년도와 t-1년도의 값을 비교하여 새로운 열을 만들거나 해당 열을 변경
# 당기매출액 / 전기매출액
# SGI
for company in df['Name'].unique():
    company_data = df[df['Name'] == company]
    company_data['SGI'] = (company_data['매출액(천원)'] / company_data['매출액(천원)'].shift(1))
    df.loc[company_data.index, 'SGI'] = company_data['SGI']
```

```python
# t-1년도와 t년도의 값을 비교하여 새로운 열을 만들거나 해당 열을 변경
# # (전기 감가상각비 / (전기 감가상각비 + 전기유형자산)) / (당기 감가상각비 / (당기 감가상각비 + 당기유형자산))
# DEPI
for company in df['Name'].unique():
    company_data = df[df['Name'] == company]
    company_data['DEPI'] = (company_data['유무형자산상각비(천원)'].shift(1) / (company_data['유무형자산상각비(천원)'].shift(1) + company_data['유형자산(천원)'].shift(1)))
    df.loc[company_data.index, 'DEPI'] = company_data['DEPI']
```

```python
# t년도와 t-1년도의 값을 비교하여 새로운 열을 만들거나 해당 열을 변경
# (당기 판관비 / 당기 매출액) / (전기 판관비 / 전기 매출액)
# SGAI
for company in df['Name'].unique():
    company_data = df[df['Name'] == company]
    company_data['SGAI'] = (company_data['판매비와관리비(천원)'] / company_data['매출액(천원)']) / (company_data['판매비와관리비(천원)'].shift(1) / company_data['매출액(천원)'].shift(1))
    df.loc[company_data.index, 'SGAI'] = company_data['SGAI']
```

```python
# t년도와 t-1년도의 값을 비교하여 새로운 열을 만들거나 해당 열을 변경
# ((당기 LTD + 당기 유동부채) / 당기 총자산) / ((전기 LTD + 전기 유동부채) / 전기 총자산)
# LVGI
for company in df['Name'].unique():
    company_data = df[df['Name'] == company]
    company_data['LVGI'] = ((company_data['비유동부채(천원)'] + company_data['유동부채(천원)']) / company_data['총자산(천원)']) / ((company_data['비유동부채(천원)'].shift(1) + company_data['유동부채(천원)'].shift(1)) / company_data['총자산(천원)'].shift(1))
    df.loc[company_data.index, 'LVGI'] = company_data['LVGI']
```

```python
# t년도와 t-1년도의 값을 비교하여 새로운 열을 만들거나 해당 열을 변경
# (당기 영업이익 - 당기 영업현금흐름) / 당기 총자산
# TATA
for company in df['Name'].unique():
    company_data = df[df['Name'] == company]
    company_data['TATA'] = (company_data['영업이익(천원)'] - company_data['영업활동으로인한현금흐름(천원)']) / company_data['총자산(천원)']
    df.loc[company_data.index, 'TATA'] = company_data['TATA']
```
